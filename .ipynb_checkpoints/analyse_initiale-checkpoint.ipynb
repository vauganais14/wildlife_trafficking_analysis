```python
import pandas as pd
import matplotlib.pyplot as plt

# Optionnel : affichage inline de matplotlib
%matplotlib inline

# Chargement des données
df = pd.read_csv("data/cites_clean.csv")

# Aperçu
df.head()
```

```python
top_taxa = df['Taxon'].value_counts().head(20)

# Plot
top_taxa.plot(kind='barh', figsize=(10, 8), title="Top 20 espèces échangées")
plt.gca().invert_yaxis()  # pour que l'espèce la plus échangée soit en haut
plt.xlabel("Nombre de transactions")
plt.tight_layout()
plt.show()
```

```python
top_exporters = df['Exporter'].value_counts().head(10)

# Plot
top_exporters.plot(kind='bar', figsize=(8, 6), title="Top 10 pays exportateurs")
plt.ylabel("Nombre de transactions")
plt.xlabel("Pays exportateur")
plt.tight_layout()
plt.show()
```

```python
top_importers = df['Importer'].value_counts().head(10)

# Plot
top_importers.plot(kind='bar', figsize=(8, 6), title="Top 10 pays importateurs", color='orange')
plt.ylabel("Nombre de transactions")
plt.xlabel("Pays importateur")
plt.tight_layout()
plt.show()
```

```python
# Conversion au cas où certaines années ne seraient pas numériques
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Groupement par année
df.groupby('Year').size().plot(kind='line', marker='o', figsize=(10, 6), title="Nombre de transactions par an")

plt.ylabel("Nombre de transactions")
plt.xlabel("Année")
plt.grid(True)
plt.tight_layout()
plt.show()
```
