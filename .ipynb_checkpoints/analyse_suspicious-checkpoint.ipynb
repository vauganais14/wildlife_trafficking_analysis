```python
import pandas as pd
import matplotlib.pyplot as plt

# Pour un affichage inline
%matplotlib inline

# Chargement du fichier suspect
df = pd.read_csv("data/suspicious_trades.csv")

# Aper√ßu
df.head()
```

```python
print(f"Nombre total de transactions suspectes : {len(df)}")
```

```python
top_taxa = df['Taxon'].value_counts().head(20)

top_taxa.plot(kind='barh', figsize=(10, 8), title="Top 20 esp√®ces impliqu√©es dans des transactions suspectes")
plt.gca().invert_yaxis()
plt.xlabel("Nombre de transactions suspectes")
plt.tight_layout()
plt.show()
```

```python
top_exporters = df['Exporter'].value_counts().head(10)

top_exporters.plot(kind='bar', figsize=(8, 6), title="Top 10 pays exportateurs impliqu√©s dans des transactions suspectes")
plt.ylabel("Nombre de transactions suspectes")
plt.xlabel("Pays exportateur")
plt.tight_layout()
plt.show()
```

```python
top_importers = df['Importer'].value_counts().head(10)

top_importers.plot(kind='bar', figsize=(8, 6), title="Top 10 pays importateurs impliqu√©s dans des transactions suspectes", color='orange')
plt.ylabel("Nombre de transactions suspectes")
plt.xlabel("Pays importateur")
plt.tight_layout()
plt.show()
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

df.groupby('Year').size().plot(kind='line', marker='o', figsize=(10, 6), title="Nombre de transactions suspectes par an")
plt.ylabel("Nombre de transactions suspectes")
plt.xlabel("Ann√©e")
plt.grid(True)
plt.tight_layout()
plt.show()
```
